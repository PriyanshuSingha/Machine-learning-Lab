```python
# q4
import pandas as pd
data={'marks':[70,75,80,85,90,75]}
df=pd.DataFrame(data)
mean_val=df['marks'].mean()
median_val=df['marks'].median()
mode_val=df['marks'].mode()
variance_val=df['marks'].var()
std_dev=df['marks'].std()
print(mean_val)
print(median_val)
print(mode_val)
print(variance_val)
print(std_dev)
```

```python
#5
import pandas as pd
data1={'id':[1,2,3,4],'marks':[80,85,56,90]}
data2={'id':[1,2,3,4],'name':['amit','riya','sohan','neha']}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
print(df1)
reshaped=df1['marks'].values.reshape(4,1)
print(reshaped)
filtered=df1[df1['marks']>85]
print(filtered)
merged=pd.merge(df1,df2,on='id')
print(merged)
df1_filled=df1.fillna(df1['marks'].mean())
print(df1_filled)
df1_filled["marks_normalized"]=(df1_filled['marks']-df1_filled['marks'].min())/(df1_filled['marks'].max()-df1_filled['marks'].min())
print(df1_filled)
```
